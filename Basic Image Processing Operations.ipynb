# Working With OpenCV
## Basic Image Processing Operations

```python
# -*- coding: utf-8 -*-
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Load an image
original_image = cv.imread('image.jpeg')

# print the shape of the image
print(original_image.shape)

# Display the image
# cv.imshow('Image', image)
# cv.waitKey(0) #Waits until a key is pressed.
# cv.destroyAllWindows()


# Convert the image from BGR to RGB
original_image = cv.cvtColor(original_image, cv.COLOR_BGR2RGB)
plt.imshow(original_image)  # Display the image using matplotlib
plt.title('Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis
```

## 1. Color Space Conversion:

```python
# Convert the image to grayscale
# This is useful for many image processing tasks
# such as edge detection, thresholding, etc.
# Grayscale images have only one channel
# instead of three (BGR)

image = cv.imread('image.jpeg')
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Display the grayscale image
plt.figure()  # Create a new figure
plt.subplot(1, 2, 1)  # Create a subplot with 1 row and 2 columns
plt.imshow(original_image)  # Display the grayscale image
plt.title('Original Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis

plt.subplot(1, 2, 2)  # Create a subplot with 1 row and 2 columns
plt.imshow(gray_image,  cmap='gray')  # Display the image using matplotlib
plt.title('Grayscale Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis
```

## 2. Image Resizing: 

```python
# Resize the image
# This is useful for reducing the size of an image
# to fit a specific size or for faster processing
# The resize function takes the image and a tuple
# representing the new size (width, height)
# The aspect ratio is not preserved in this case
# To preserve the aspect ratio, you need to calculate
# the new size based on the original size and the
# desired width or height
# For example, if you want to resize the image
# to a width of 200 pixels and preserve the aspect ratio,
# you can calculate the new height as follows:
# new_height = int(original_height * (200 / original_width))
# Here is an example of resizing the image to a fixed size
# of 200x150 pixels
# Note: This may distort the image if the aspect ratio
# is not preserved
# Load the image again
# and resize it to a fixed size
# of 200x150 pixels
# Note: This may distort the image if the aspect ratio
# is not preserved
# Load the image again
# and resize it to a fixed size
# of 200x150 pixels
# Note: This may distort the image if the aspect ratio
# is not preserved
# Load the image again
# and resize it to a fixed size
# of 200x150 pixels

image = cv.imread('image.jpeg')
resized_image = cv.resize(image, (200, 100)) # Resize to 200x150 pixels
# Display the grayscale image
plt.figure()  # Create a new figure
plt.subplot(1, 2, 1)  # Create a subplot with 1 row and 2 columns
plt.imshow(original_image)  # Display the grayscale image
plt.title('Original Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis

plt.subplot(1, 2, 2)  # Create a subplot with 1 row and 2 columns
plt.imshow(cv.cvtColor(resized_image, cv.COLOR_BGR2RGB))  # Display the image using matplotlib
plt.title('Resized Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis
```

## 3. Image Cropping: 

```python
# Crop the image
# Cropping is the process of removing unwanted
# parts of an image
# You can crop an image by specifying the
# coordinates of the top-left and bottom-right
# corners of the rectangle you want to keep
# The coordinates are specified as (y1, y2, x1, x2)
# where y1 and y2 are the row indices (height)
# and x1 and x2 are the column indices (width)

image = cv.imread('image.jpeg')
cropped_image = image[50:, 150:] # Crop from row 50 to 149, and column 100 to 249

# Display the original image
plt.figure()  # Create a new figure
plt.subplot(1,2,1)  # Create a subplot with 1 row and 2 columns
plt.imshow(original_image)  # Display the original image
plt.title('Original Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis

# Display the cropped image
plt.subplot(1,2,2)  # Create a subplot with 1 row and 2 columns
plt.imshow(cv.cvtColor(cropped_image, cv.COLOR_BGR2RGB))  # Display the cropped image
plt.title('Cropped Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis
```

## 4. Image Filtering (Blurring): 

```python
# Apply Gaussian blur
# Gaussian blur is a low-pass filter that smooths
# the image by averaging the pixel values in a
# Gaussian distribution
# It is useful for reducing noise and detail
# in an image
# The Gaussian blur function takes the image,
# the size of the kernel (width and height),
# and the standard deviation of the Gaussian
# distribution as arguments

image = cv.imread('image.jpeg')
blurred_image = cv.GaussianBlur(image, (5, 5), 0) # Apply a 5x5 Gaussian blur

# display the original image
plt.figure()  # Create a new figure
plt.subplot(1, 2, 1)  # Create a subplot with 1 row and 2 columns
plt.imshow(original_image)  # Display the original image
plt.title('Original Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis

# display the blurred image
plt.subplot(1, 2, 2)  # Create a subplot with 1 row and 2 columns
plt.imshow(cv.cvtColor(blurred_image, cv.COLOR_BGR2RGB))  # Display the blurred image
plt.title('Blurred Image')  # Set the title of the plot
plt.axis('off')  # Hide the axis
```
